```python
# ---- Task 01: Bar chart / Histogram ----
# Objective: Visualize the distribution of a categorical or continuous variable
# Example: Distribution of population across countries in the year 2020

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset 
file_path = (r"C:\Users\Tamilarasan K\Downloads\Dataset_1 - Prodigy task 1.xlsx")
df = pd.read_excel(file_path, skiprows=3)

# Step 2: Check columns
print("Available columns:\n", df.columns)

# Step 3: Select the year column (example: 2020)
year = "2020"

# Step 4: Drop missing values for that year
data = df[year].dropna()

# Step 5: Create histogram
plt.figure(figsize=(8,5))
plt.hist(data, bins=20, color="skyblue", edgecolor="black")
plt.title(f"Distribution of Population in {year}")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# ---- OPTION B: If you want bar chart for a categorical variable ----
# Example: Top 10 countries by population in 2020
top_countries = df[['Country Name', year]].dropna().nlargest(10, year)

plt.figure(figsize=(10,6))
plt.bar(top_countries['Country Name'], top_countries[year], color="orange", edgecolor="black")
plt.title(f"Top 10 Countries by Population in {year}")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha="right")
plt.savefig("population_histogram.png")
plt.savefig("population_top10.png")
plt.show()
```
